```python
signal_all=np.vstack((signal_Tol,signal_Ace,signal_M,signal_CO,signal_Meth,signal_Eth,signal_But,signal_Benz,signal_Am,signal_Ac))

signal_all_r=np.reshape(signal_all,(len(signal_all),(6*26000)))

print(np.shape(signal_all_r))
```

```python
# SVM Classifier (RBF Kernel)

from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix


y=np.hstack((np.zeros(20),np.ones(20),np.ones(20)*2,np.ones(20)*3,np.ones(20)*4,np.ones(20)*5,np.ones(20)*6,np.ones(20)*7,np.ones(20)*8,np.ones(20)*9))

accuracy_over_splits=[]

cm_over_splits=[]

for i in range(0,50):
    X_train,X_test,y_train,y_test=train_test_split(signal_all_r,y,train_size=0.8,test_size=0.2,stratify=y,random_state=i)
    
    svm_model=OneVsRestClassifier(SVC(kernel='rbf')).fit(X_train,y_train)

    svm_predictions=svm_model.predict(X_test)

    accuracy=svm_model.score(X_test,y_test)

    cm=confusion_matrix(y_test,svm_predictions)
    
    accuracy_over_splits.append(accuracy)
    
    cm_over_splits.append(cm)
```

```python
plt.hist(accuracy_over_splits)

plt.xlabel('Accuracy')

plt.ylabel('Number of splits')

plt.grid(True)
```

```python
# SVM Classifier (Linear Kernel)

from sklearn.model_selection import train_test_split
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix


y=np.hstack((np.zeros(20),np.ones(20),np.ones(20)*2,np.ones(20)*3,np.ones(20)*4,np.ones(20)*5,np.ones(20)*6,np.ones(20)*7,np.ones(20)*8,np.ones(20)*9))

accuracy_over_splits_lin=[]

cm_over_splits_lin=[]

for i in range(0,50):
    X_train,X_test,y_train,y_test=train_test_split(signal_all_r,y,train_size=0.8,test_size=0.2,stratify=y,random_state=i)
    
    svm_model=OneVsRestClassifier(SVC(kernel='linear')).fit(X_train,y_train)

    svm_predictions=svm_model.predict(X_test)

    accuracy=svm_model.score(X_test,y_test)

    cm=confusion_matrix(y_test,svm_predictions)
    
    accuracy_over_splits_lin.append(accuracy)
    
    cm_over_splits_lin.append(cm)
```

```python
plt.hist(accuracy_over_splits_lin)

plt.xlabel('Accuracy')

plt.ylabel('Number of splits')

plt.grid(True)
```

```python
# Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier

accuracy_over_splits_decision_tree=[]

cm_over_splits_decision_tree=[]

for i in range(0,50):
    X_train,X_test,y_train,y_test=train_test_split(signal_all_r,y,train_size=0.8,test_size=0.2,stratify=y,random_state=i)
    
    dtree_model=DecisionTreeClassifier(max_depth=2).fit(X_train,y_train)

    dtree_predictions=dtree_model.predict(X_test)

    accuracy_dtree=dtree_model.score(X_test,y_test)

    cm_dtree=confusion_matrix(y_test,dtree_predictions)
    
    accuracy_over_splits_decision_tree.append(accuracy_dtree)
    
    cm_over_splits_decision_tree.append(cm_dtree)
    
```

```python
plt.hist(accuracy_over_splits_decision_tree)

plt.xlabel('Accuracy')

plt.ylabel('Number of splits')

plt.grid(True)
```

```python
# K Nearest Neighbor Classifier

from sklearn.neighbors import KNeighborsClassifier

accuracy_over_splits_knn=[]

cm_over_splits_knn=[]

for i in range(0,50):
    X_train,X_test,y_train,y_test=train_test_split(signal_all_r,y,train_size=0.8,test_size=0.2,random_state=i,stratify=y)
    
    knn_model=KNeighborsClassifier(n_neighbors=4).fit(X_train,y_train)

    accuracy_knn=knn_model.score(X_test,y_test)
    
    knn_predictions=knn_model.predict(X_test)

    cm_knn=confusion_matrix(y_test,knn_predictions)
    
    accuracy_over_splits_knn.append(accuracy_knn)
    
    cm_over_splits_knn.append(cm_knn)
```

```python
plt.hist(accuracy_over_splits_knn)

plt.xlabel('Accuracy')

plt.ylabel('Number of splits')

plt.grid(True)
```

```python
accuracy_classifiers=[accuracy_over_splits_lin,accuracy_over_splits,accuracy_over_splits_decision_tree,accuracy_over_splits_knn]

fig=plt.figure(figsize=(10,7))

ax=fig.add_axes([0,0,1,1])

bp=ax.boxplot(accuracy_classifiers)

ax.set_xticklabels(['SVM (Linear)','SVM (RBF)','Decision Tree (depth=2)', 'K-Nearest Neighbors (K=4)'])

ax.set_xlabel('Classifiers')

ax.set_ylabel('Accuracy')

```
